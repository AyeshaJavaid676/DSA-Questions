```python
def isPalindrome(s):

    cleaned = "".join(char.lower() for char in s if char.isalnum())

    L = 0
    R = len(cleaned) - 1

    while L < R:
        if cleaned[L] != cleaned[R]:
            return False

        L += 1
        R -= 1

    return True
```

```python
result = isPalindrome("A man, a plan, a canal: Panama")
print(result)
```
