# **Top K frequent Elements**

```python
def topKFrequent(nums, k):
    count_map = {}
    for n in nums:
        count_map[n] = 1 + count_map.get(n, 0)

    flipped_list = []
    for num, count in count_map.items():
        flipped_list.append([count, num])

    flipped_list.sort()

    result = []
    for i in range(1, k + 1):
        result.append(flipped_list[-i][1])

    return result
```

```python
def topKFrequent(nums, k):
    # 1. Your logic: Count frequencies
    count_map = {}
    for n in nums:
        count_map[n] = 1 + count_map.get(n, 0)

    # 2. Sort the dictionary by its VALUES (the counts)
    # We want the highest counts first, so we use reverse=True
    sorted_items = sorted(count_map.items(), key=lambda item: item[1], reverse=True)

    # 3. Take the first 'k' elements and return their 'keys'
    result = []
    for i in range(k):
        result.append(sorted_items[i][0])

    return result
```
